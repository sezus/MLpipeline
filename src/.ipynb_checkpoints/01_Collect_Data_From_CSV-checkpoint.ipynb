```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
def anomaly_normalize(df,renamecolumns,feature_fill_list,feature_fill_list2,normalize_feature_list):
    #Rename Columns:
    print(renamecolumns)
    df=df.rename(columns=renamecolumns)
    #Remove Anomaly valuesColumns:
    df=remove_anomaly_values(df,feature_fill_list,uplimit=100,lowlimit=-5)
    df=remove_anomaly_values(df,feature_fill_list2,lowlimit=-5)
    #Fill NA Empty Rows:
    df=fillna(df)
    df_before_normal=df.copy()
    #Normalize all continious columns burda sadece verilen list ile verilen kolonlarÄ± normalize eder
    df=normalize_all(df,normalize_feature_list)
    return df,df_before_normal
```

```python
def upload_multiple_sheets(Path_list,File_list,Sheet_list,range_list,refcolumn):
    combined_df=[]
    print(Path_list)
    print(File_list)
    print(Sheet_list)
    print(range_list)
    for i in range(len(Path_list)):         
        print(range_list[i])
        df=upload_single_sheet(Path_list[i],File_list[i],Sheet_list[i],range_list[i])    
        df=remove_key_columns_duplicate_rows(df,refcolumn)
        combined_df.append(df)   
    df=combined_df[0]
    for i in range(1,len(combined_df)):
        #print(df.columns)
        df=df.merge(combined_df[i].set_index(refcolumn),on=refcolumn,how='left')   
    return df
def upload_all_sheets_single_file(path,filename,skiprows=None):
    """
    Parameters
    ----------
    obj : scalar or array-like
        Object to check for null or missing values.

    Returns
    -------
    bool or array-like of bool
        For scalar input, returns a scalar boolean.
        For array input, returns an array of boolean indicating whether each
        corresponding element is missing.    
    """
    print(path)
    print(filename)
    #if len(path)
    addres=path+'/' +filename
    print(addres)
    if skiprows==None:
        alls=pd.read_excel(addres,sheet_name=None) 
    else:
        alls=pd.read_excel(addres,sheet_name=None,skiprows=skiprows) 
        
    result=[]
    for i in alls:
        #result=result.append(alls[i], sort=False)
        alls[i]=remove_all_columns_duplicate_rows(alls[i])
        alls[i]=drop_full_nan_columns(alls[i])  

    #Allsheets = pd.concat(Allsheets,ignore_index=True)
    return alls

def upload_single_sheet(path,filename,sheetname,skiprows=None):
    """
    Parameters
    ----------
    obj : scalar or array-like
        Object to check for null or missing values.

    Returns
    -------
    bool or array-like of bool
        For scalar input, returns a scalar boolean.
        For array input, returns an array of boolean indicating whether each
        corresponding element is missing.    
    """
    #if len(path)
    addres=path+'/' +filename
    df=pd.read_excel(addres,sheet_name=sheetname,skiprows=range(skiprows[0],skiprows[1]))  
    
    #remove duplicate rows
    df=remove_all_columns_duplicate_rows(df)
    
    #remove nan columns
    df=drop_full_nan_columns(df)  
    
    #remove nan rows
    df=df.dropna(how='all')
    
    return df

def drop_full_nan_columns(df):    
    for i in df.columns:
        if df[i].isnull().all()==True :
            df=df.drop(i,axis='columns')        
    return df

def remove_all_columns_duplicate_rows(df):    
    df=df[~df.duplicated()]     
    return df

def remove_key_columns_duplicate_rows(df,refcolumn):
    df=df[~df[refcolumn].duplicated()] 
    return df    
def remove_anomaly_values(df,feature_list,uplimit=None,lowlimit=None):
    for i in feature_list:
        if uplimit!=None:
            df.loc[df[i]>uplimit,i]=np.nan
        if lowlimit!=None:
            df.loc[df[i]<lowlimit,i]=np.nan 
    return df


def fill_feature_nan_values(df,feature_list,method):
    for i in feature_list:
        df[i]=df[i].fillna(method=method)
    return df

def fillna(df):      
    features=pd.DataFrame(df.isnull().sum(),columns=list('A'))
    for i in list( range(0, features.size)):     
        #if features.iat[i,0] is not '0':
        df[features.index[i]]=df[features.index[i]].fillna(method='ffill')
        df[features.index[i]]=df[features.index[i]].fillna(method='bfill')
    return(df)       
def normalize(val):
    val=(val-np.min(val))/(np.max(val)-np.min(val))
    return val
def normalize_all(df,feature_list):    
    dfn=df.copy()    
    for i in feature_list:
        dfn[i]=normalize(dfn[i].values)
    return dfn
def normalize_all_features(df):
    dfn=df.copy()
    feature_list=dfn.dtypes[dfn.dtypes=='float64'].index    
    for i in feature_list:
        dfn[i]=normalize(dfn[i].values)
    return dfn
def combine_df_Period(df1,refcolumn1,df2,refcolumn2):
    df1[refcolumn1[1]]=0
    for i in range(df2[refcolumn2].shape[0]):
        startT02=df1.loc[df1[refcolumn1[0]]> df2[refcolumn2[0]][i]]
        #print(startT02)
        df1.loc[startT02[startT02[refcolumn1[0]]<df2[refcolumn2[1]][i]].index,refcolumn1[1]] = df2[refcolumn2[2]][i]         
    #print(df1.Events.unique())
    return df1
from sklearn import preprocessing
def create_binary(df,refcolumn):
    binary=refcolumn+'_binary'
    df[binary]=df[refcolumn].copy()
    df.loc[df[refcolumn]!=0,binary]=1
    return df
def create_multiclass(df,refcolumn):
    multiclass=refcolumn+'_multiclass'
    le = preprocessing.LabelEncoder()
    df[refcolumn]=df[refcolumn].astype(str).copy()
    le.fit(df[refcolumn])
    df[multiclass]=le.transform(df[refcolumn])    
    return df
def create_onehot(df,refcolumn):
    ref=df[refcolumn]
    df=pd.get_dummies(df, columns=[refcolumn])
    df[refcolumn]=ref
    return df
from sklearn import preprocessing
def create_binary_multiclass_onehot_labels(df,refcolumn):
    df=create_binary(df,refcolumn)
    df=create_multiclass(df,refcolumn)
    df=create_onehot(df,refcolumn)
    return df
def create_component_labels(df,refcolumn,faultdf,faultrefcolumn,faultcategory):
    component=faultdf[faultcategory].unique()    
    df['Components']=df[refcolumn]
    for i in component:
        #print(i)
        #print(refcolumn)
        #Componentlere Ait Eventleri cikar
        ComponentEvents=faultdf[faultdf[faultcategory]==i][faultrefcolumn].unique()
        #Componentlere Ait Eventleri cikar
        df[i]=0
        df.loc[df[refcolumn].isin(ComponentEvents),i]=1
        #print(df[i].unique())
        #print(ComponentEvents)
        Even=i+'Ev'
        df[Even]=0
        df.loc[df[refcolumn].isin(ComponentEvents),Even]=df.loc[df[refcolumn].isin(ComponentEvents)].Events
        create_multiclass(df,Even)
        df.loc[df[refcolumn].isin(ComponentEvents),'Components']=i
    create_multiclass(df,'Components')
    return df

```

```python
##BurayÄ± configuratif yap
def fault(dffault,path):
    ## excel oku
    Faults_MainCategory = pd.read_excel(path)
    ## Gereksiz ilk kolonu drop et.
    Faults_MainCategory=Faults_MainCategory.drop(Faults_MainCategory.columns[0], axis='columns')
    ## kolon isimlerini deÄiÅtir
    Faults_MainCategory.columns=['Number', 'Braking', 'Main Category', 'Event ID']
    ## Fault(Target) deÄiÅken ile Event listesini merge et
    FaultAllCategoryEvents = pd.merge(dffault,Faults_MainCategory[['Main Category', 'Event ID']],how='left',  on='Event ID')
    ## na olan kÄ±sÄ±mlarÄ± drop et.
    FaultAllCategoryEvents=FaultAllCategoryEvents[FaultAllCategoryEvents['Main Category'].notna()]
    return FaultAllCategoryEvents
```
