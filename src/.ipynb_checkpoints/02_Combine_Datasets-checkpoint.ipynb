```python
import os
import yaml
import pandas as pd
import numpy as np
import pickle

## count according to numberof reference column (sum,mean,median ya da tanımlanan bir fonksiyon input olarak verilebilir)
def select_WT_fault_count_bigger_than_threshold(df,numbercolumn,labelcolumn,threshold):
    tagnumber=[]
    for i in df[numbercolumn].unique():
        #print(i)
        systemnumber=df[df[numbercolumn]==i].copy()
        #print(systemnumber[labelcolumn].sum())
        if systemnumber[labelcolumn].sum()>threshold:  
            tagnumber.append(i)
    return tagnumber

# Function to load yaml configuration file
def load_config(config_name):
    with open(os.path.join(CONFIG_PATH,  config_name), 'r', encoding='utf8') as file:
        config = yaml.safe_load(file)

    return config

def concat_data_set(datalist):
    for data in datalist:
        concatdata = pd.concat(datalist,ignore_index=True)
    return concatdata

#Convert dataframe to list according to reference column unique values
#türbin datalarını birleştirmek üzere kullanılır. 
def convert_dataframe_to_list(df,refcolumn):
    X_list=[]
    for i in df[refcolumn].unique():
        ref_df=df[df[refcolumn]==i].copy()
        ref_df.drop(refcolumn,axis='columns')
        X_list.append(ref_df) 
    return X_list


if __name__ == '__main__':
    CONFIG_PATH = "."
    config = load_config('importcsv.yaml')
    Upload_tag=config["Upload_list_all"]["Upload_Tags"]
    tag=config["Upload_list_all"][Upload_tag[0]]
    Tag_Column=config["Upload_list_all"]["Tag_Column"]
    raw_path=config["Upload_list_all"]['Parent_path']+"data/01_raw/" 
    upload_path=raw_path+"uploaded/" 
    intermediate_path=config["Upload_list_all"]['Parent_path']+"/data/02_intermediate/" 
    combined_path=intermediate_path + "CombinedDatasets/"
    Target_Label=config["Upload_list_all"]["Target_Column"]
    

    Features_List=[]
    Labels_List=[]
    for tag in Upload_tag:
        with open(intermediate_path+'Features'+tag, 'rb') as config_file:
            tag_f= pickle.load(config_file)
        with open(intermediate_path +'Label'+tag, 'rb') as config_file:
            tag_i= pickle.load(config_file)

        tag_f[Tag_Column]=tag
        tag_i[Tag_Column]=tag
        Features_List.append(tag_f)
        Labels_List.append(tag_i)     

    Features_df=concat_data_set(Features_List)
    Labels_df=concat_data_set(Labels_List)
    #print(Features_List[0].columns)
    #print(Features_df.columns)
    selectedLabels=select_WT_fault_count_bigger_than_threshold(Labels_df,Tag_Column,Target_Label,100)
    Features_selected=Features_df[Features_df[Tag_Column].isin(selectedLabels)]
    #Labels_selected=Labels_df[Labels_df[Tag_Column].isin(selectedLabels)]
    Labels_df=Labels_df[Labels_df[Tag_Column].isin(selectedLabels)]
    Labels_selected=Labels_df[[Target_Label,Tag_Column]]

    Features_selected_list=[]
    Labels_selected_list=[]
    
    for i in Features_List:
        select=i[i.TURBINE_NUMBER.isin(selectedLabels)].copy()
        #select=select[select.columns[1:-2]]
        Features_selected_list.append(select)

    for i in Labels_List:
        i_df=i[i.TURBINE_NUMBER.isin(selectedLabels)].copy()
        print(i.columns)
        select=i_df[[Target_Label]]
        Labels_selected_list.append(select)
        
        
    #print(Labels_selected_list)
    #print(Labels_selected.columns)

    with open(combined_path + 'Features_df', 'wb') as config_file:
        pickle.dump(Features_selected, config_file)
    with open(combined_path + 'Labels_df', 'wb') as config_file:
        pickle.dump(Labels_selected, config_file)

    with open(combined_path + 'Features_List', 'wb') as config_file:
        pickle.dump(Features_selected_list, config_file)
    with open(combined_path + 'Labels_List', 'wb') as config_file:
        pickle.dump(Labels_selected_list, config_file)
```
