```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D  

%matplotlib inline
```

```python
with open('method_k_dict.dictionary', 'rb') as config_dictionary_file:
    FS_dict = pickle.load(config_dictionary_file)
```

```python
FS_dict_Chi2 = FS_dict['chi2']
FS_dict_Fval = FS_dict['f_classif']
FS_dict_MI = FS_dict['mutual_info_classif']
```

```python
f1_train_Chi2 = []
f1_valid_Chi2 = []
f1_train_Fval = []
f1_valid_Fval = []
f1_train_MI = []
f1_valid_MI = []
for i in range(1,32):
    f1_train_Chi2.append(FS_dict_Chi2[str(i)]['f1_train'])
    f1_valid_Chi2.append(FS_dict_Chi2[str(i)]['f1_valid'])
    f1_train_Fval.append(FS_dict_Fval[str(i)]['f1_train'])
    f1_valid_Fval.append(FS_dict_Fval[str(i)]['f1_valid'])
    f1_train_MI.append(FS_dict_MI[str(i)]['f1_train'])
    f1_valid_MI.append(FS_dict_MI[str(i)]['f1_valid'])
```

```python
fig, axes = plt.subplots(nrows = 3)
fig.set_figheight(12)
fig.set_figwidth(10)
fig.tight_layout(pad=5.0)
#plt.style.use('seaborn')
fig.legend(['f1_train','f1_validation'])
axes[0].plot(range(1, 32), f1_train_Chi2)
axes[0].plot(range(1, 32), f1_valid_Chi2)
axes[0].set_ylim(0,1)
axes[0].set_title('Chi Square')
axes[0].set(xlabel='number of features', ylabel='f1 score')
axes[0].grid()
axes[0].legend(['train','validation'])
axes[0].set_ylim([0.8,1])
axes[1].plot(range(1, 32), f1_train_Fval)
axes[1].plot(range(1, 32), f1_valid_Fval)
axes[1].set_ylim(0,1)
axes[1].set_title('ANOVA F-value')
axes[1].set(xlabel='number of features', ylabel='f1 score')
axes[1].grid()
axes[1].legend(['train','validation'])
axes[1].set_ylim([0.8,1])
axes[2].plot(range(1, 32), f1_train_MI)
axes[2].plot(range(1, 32), f1_valid_MI)
axes[2].set_ylim(0,1)
axes[2].set_title('Mutual Information')
axes[2].set(xlabel='number of features', ylabel='f1 score')
axes[2].grid()
axes[2].legend(['train','validation'])
axes[2].set_ylim([0.8,1])
```

```python
with open('bests_XGBoost.dictionary', 'rb') as config_dictionary_file:
    bests_XGBoost = pickle.load(config_dictionary_file)
with open('bests_RFC.dictionary', 'rb') as config_dictionary_file:
    bests_RFC = pickle.load(config_dictionary_file)
with open('bests_LGBM.dictionary', 'rb') as config_dictionary_file:
    bests_LGBMC = pickle.load(config_dictionary_file)
with open('bests_CatBoost.dictionary', 'rb') as config_dictionary_file:
    bests_catboost = pickle.load(config_dictionary_file)
```

```python
param_XGBoost_max_depth = bests_XGBoost['cv_results']['param_max_depth'].data
param_RFC_max_depth = bests_RFC['cv_results']['param_max_depth'].data
param_LGBMC_max_depth = bests_LGBMC['cv_results']['param_max_depth'].data
param_catboost_depth = bests_catboost['cv_results']['param_depth'].data
param_XGBoost_min_child_weight = bests_XGBoost['cv_results']['param_min_child_weight'].data
param_RFC_min_impurity_split = bests_RFC['cv_results']['param_min_impurity_split'].data
param_LGBMC_min_child_weight = bests_LGBMC['cv_results']['param_min_child_weight'].data
param_catboost_l2_leaf_reg = bests_catboost['cv_results']['param_l2_leaf_reg'].data
```

```python
scores_XGBoost = bests_XGBoost['cv_results']['mean_test_score']
scores_RFC = bests_RFC['cv_results']['mean_test_score']
scores_LGBMC = bests_LGBMC['cv_results']['mean_test_score']
scores_catboost = bests_catboost['cv_results']['mean_test_score']
```

```python
param_XGBoost_min_child_weight_ = []
for i in range(len(param_XGBoost_min_child_weight)):
    if param_XGBoost_min_child_weight[i] == 0.01:
        param_XGBoost_min_child_weight_.append(-2)
    if param_XGBoost_min_child_weight[i] == 0.1:
        param_XGBoost_min_child_weight_.append(-1)
    if param_XGBoost_min_child_weight[i] == 1:
        param_XGBoost_min_child_weight_.append(0)
    if param_XGBoost_min_child_weight[i] == 10:
        param_XGBoost_min_child_weight_.append(1)
    if param_XGBoost_min_child_weight[i] == 100:
        param_XGBoost_min_child_weight_.append(2)
```

```python
plt.style.use('classic')
fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(10)
ax = fig.add_subplot(111,projection='3d')
ax.scatter(param_XGBoost_max_depth, param_XGBoost_min_child_weight_, scores_XGBoost, marker = "^", s=30, c = scores_XGBoost)
ax.scatter(param_XGBoost_max_depth[scores_XGBoost.argmax()], param_XGBoost_min_child_weight_[scores_XGBoost.argmax()], max(scores_XGBoost), marker = "o", s=100, c = 'black')
ax.set_xlabel('max_depth')
ax.set_ylabel('min_child_weight(log scale)')
ax.set_zlabel('mean validation f1 score')
ax.view_init(30, 140)
ax.set_title('XGBoosting')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)

text = ax.text(param_XGBoost_max_depth[scores_XGBoost.argmax()], 
               param_XGBoost_min_child_weight_[scores_XGBoost.argmax()], 
               max(scores_XGBoost)+0.04, 
               'max_depth = '+str(param_XGBoost_max_depth[scores_XGBoost.argmax()])+
              '\nmin_child_weight = '+str(param_XGBoost_min_child_weight[scores_XGBoost.argmax()])+
              '\nmean validation score = '+str(max(scores_XGBoost)),
              [0,0,0], bbox=props)
```

```python
param_RFC_min_impurity_split_ = []
for i in range(len(param_RFC_min_impurity_split)):
    if param_RFC_min_impurity_split[i] == 1e-1:
        param_RFC_min_impurity_split_.append(-1)
    if param_RFC_min_impurity_split[i] == 1e-2:
        param_RFC_min_impurity_split_.append(-2)
    if param_RFC_min_impurity_split[i] == 1e-3:
        param_RFC_min_impurity_split_.append(-3)
    if param_RFC_min_impurity_split[i] == 1e-4:
        param_RFC_min_impurity_split_.append(-4)
    if param_RFC_min_impurity_split[i] == 1e-5:
        param_RFC_min_impurity_split_.append(-5)
```

```python
plt.style.use('classic')
fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(10)
ax = fig.add_subplot(111,projection='3d')
ax.scatter(param_RFC_max_depth, param_RFC_min_impurity_split_, scores_RFC, marker = "^", s=30, c = scores_RFC)
ax.scatter(param_RFC_max_depth[scores_RFC.argmax()], param_RFC_min_impurity_split_[scores_RFC.argmax()], max(scores_RFC), marker = "o", s=100, c = 'black')
ax.set_xlabel('max_depth')
ax.set_ylabel('min_impurity_split (log scale)')
ax.set_zlabel('mean validation f1 score')
ax.view_init(30, 140)
ax.set_title('Random Forest Classifier')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)

text = ax.text(param_RFC_max_depth[scores_RFC.argmax()], 
               param_RFC_min_impurity_split_[scores_RFC.argmax()], 
               max(scores_RFC)+0.04, 
               'max_depth = '+str(param_RFC_max_depth[scores_RFC.argmax()])+
              '\nmin_child_weight = '+str(param_RFC_min_impurity_split[scores_RFC.argmax()])+
              '\nmean validation score = '+str(max(scores_RFC)),
              [0,0,0], bbox=props)
```

```python
param_LGBMC_min_child_weight_ = []
for i in range(len(param_LGBMC_min_child_weight)):
    if param_LGBMC_min_child_weight[i] == 0.01:
        param_LGBMC_min_child_weight_.append(-2)
    if param_LGBMC_min_child_weight[i] == 0.1:
        param_LGBMC_min_child_weight_.append(-1)
    if param_LGBMC_min_child_weight[i] == 1:
        param_LGBMC_min_child_weight_.append(0)
    if param_LGBMC_min_child_weight[i] == 10:
        param_LGBMC_min_child_weight_.append(1)
    if param_LGBMC_min_child_weight[i] == 100:
        param_LGBMC_min_child_weight_.append(2)
```

```python
plt.style.use('classic')
fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(10)
ax = fig.add_subplot(111,projection='3d')
ax.scatter(param_LGBMC_max_depth, param_LGBMC_min_child_weight_, scores_LGBMC, marker = "^", s=30, c = scores_LGBMC)
ax.scatter(param_LGBMC_max_depth[scores_LGBMC.argmax()], param_LGBMC_min_child_weight_[scores_LGBMC.argmax()], max(scores_LGBMC), marker = "o", s=100, c = 'black')
ax.set_xlabel('max_depth')
ax.set_ylabel('min_child_weight (log scale)')
ax.set_zlabel('mean validation f1 score')
ax.view_init(30, 140)
ax.set_title('LGBM Classifier')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)

text = ax.text(param_LGBMC_max_depth[scores_LGBMC.argmax()], 
               param_LGBMC_min_child_weight_[scores_LGBMC.argmax()], 
               max(scores_LGBMC)+0.01, 
               'max_depth = '+str(param_LGBMC_max_depth[scores_LGBMC.argmax()])+
              '\nmin_child_weight = '+str(param_LGBMC_min_child_weight[scores_LGBMC.argmax()])+
              '\nmean validation score = '+str(max(scores_LGBMC)),
              [0,0,0], bbox=props)
```

```python
param_catboost_l2_leaf_reg_ = []
for i in range(len(param_catboost_l2_leaf_reg)):
    if param_catboost_l2_leaf_reg[i] == 0.01:
        param_catboost_l2_leaf_reg_.append(-2)
    if param_catboost_l2_leaf_reg[i] == 0.1:
        param_catboost_l2_leaf_reg_.append(-1)
    if param_catboost_l2_leaf_reg[i] == 1:
        param_catboost_l2_leaf_reg_.append(0)
    if param_catboost_l2_leaf_reg[i] == 10:
        param_catboost_l2_leaf_reg_.append(1)
    if param_catboost_l2_leaf_reg[i] == 100:
        param_catboost_l2_leaf_reg_.append(2)
```

```python
plt.style.use('classic')
fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(10)
ax = fig.add_subplot(111,projection='3d')
ax.scatter(param_catboost_depth, param_catboost_l2_leaf_reg_, scores_catboost, marker = "^", s=30, c = scores_catboost)
ax.scatter(param_catboost_depth[scores_catboost.argmax()], param_catboost_l2_leaf_reg_[scores_catboost.argmax()], max(scores_catboost), marker = "o", s=100, c = 'black')
ax.set_xlabel('depth')
ax.set_ylabel('l2_leaf_reg (log scale)')
ax.set_zlabel('mean validation f1 score')
ax.view_init(30, 140)
ax.set_title('CatBoost Classifier')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.3)

text = ax.text(param_catboost_depth[scores_catboost.argmax()], 
               param_catboost_l2_leaf_reg_[scores_catboost.argmax()]-0.2, 
               max(scores_catboost), 
               'depth = '+str(param_catboost_depth[scores_catboost.argmax()])+
              '\nl2_leaf_reg = '+str(param_catboost_l2_leaf_reg[scores_catboost.argmax()])+
              '\nmean validation score = '+str(max(scores_catboost)),
              [0,0,0], bbox=props)
```
