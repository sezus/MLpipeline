```python
import os
import sys
sys.path.append('..') 
import yaml
import pandas as pd
import numpy as np
import pickle
from src.data.Collect_Data_From_CSV import upload_multiple_sheets,upload_single_sheet
from src.intermediate.Clean_Normalize import clean_normalize
from src.intermediate.Create_Labels import create_labels
from src.utils.utils_p import YamlParser,save_dataset_pickle,download_pickle,Config_Paths
```

```python
#####LOAD CONFIG PARSE PATHS#####
#yaml file Load path:
CONFIG_PATH = "D:/Users/sezgis/Desktop/Projects/Git/SmartPDM/Diagnostic_Pitch_System_Final/src"

#Load project config from yaml file:
yaml_parser = YamlParser(CONFIG_PATH,'importcsv.yaml')
config = yaml_parser.get_yaml_file()

#Dataset Tags:
Upload_tag=config["Upload_list_all"]["Upload_Tags"]

#Select Tag to Upload
TagName=Upload_tag[0]

#Dataset Information:
tag=config["Upload_list_all"][TagName]

#Raw Data Path
raw_path=config["Upload_list_all"]['Parent_path']+"data/01_raw/" 

#Intermediate Path
intermediate_path=config["Upload_list_all"]['Parent_path']+"/data/02_intermediate/" 

#Parent Project Path
parent_path=config["Upload_list_all"]['Parent_path'] 
```

# LOAD CSV Files

```python
#####01_Load CSV#####
for TagName in config["Upload_list_all"]["Upload_Tags"]:
    
    #Dataset Information:
    tag=config["Upload_list_all"][TagName]
    component=config["Upload_list_all"]['External_Datasets']['Component']
    
    #01_1.Load Logs:
    Logs=upload_multiple_sheets(parent_path,tag['Path_list'],tag['File_list'],tag['Sheet_list'],tag['Range_list'],tag['refcolumn'])
    #01_2.load faults:
    Faults=upload_single_sheet(parent_path,tag['Fault_path'],tag['Fault_file'],tag['Fault_sheet'],skiprows=tag['Fault_range'])  
    #01_3.Load Category:
    FaultComponent=upload_single_sheet(parent_path,component['Path'],component['File'],component['Sheet'],skiprows=component['Range'])  
    FaultComponent.drop('Unnamed: 0',axis='columns')
    
    #Save Uploaded Dataframes as Logs and Faults from Raw data:
    save_dataset_pickle(raw_path,'Logs'+TagName,Logs)
    save_dataset_pickle(raw_path,'Faults'+TagName,Faults)
    
    #External Category/Component Dataset
    save_dataset_pickle(raw_path,'FaultComponent',FaultComponent)

    #Download Pickle Files
    Logs=download_pickle(raw_path,'Logs'+TagName)
    Faults=download_pickle(raw_path,'Faults'+TagName)
    print('Logs%d:',Logs.columns)
    print('Faults %d:',Faults.columns)
    
    #####02_Clean & Normalize Logs(Features) data#####
    #Clean and Normalize 
    Logs,Logs_not_normal,Logs_Faults=clean_normalize(config,TagName,Logs,Faults)
    #Save Normalize Logs & Not Normalize Logs & Combined Dataset Logs+Faults:
    save_dataset_pickle(intermediate_path,'Logs_normal_fill' + TagName,Logs)
    save_dataset_pickle(intermediate_path,'Logs_not_normal_fill'+ TagName,Logs_not_normal)
    save_dataset_pickle(intermediate_path,'Logs_Faults' + TagName,Logs_Faults)    
    
    component=config["Upload_list_all"]['External_Datasets']['Component']

    #Download Pickle Files
    Logs_Faults=download_pickle(intermediate_path,'Logs_Faults'+TagName)
    FaultComponent=download_pickle(raw_path,'FaultComponent')
    print('Logs%d:',Logs.columns)
    print('Faults %d:',Faults.columns)
    #Event referans alarak once binary sonra multiclass sonra da onehot labellar oluştur.
    #Yeni Labellar oluşturmadan önceki kolon sayısı before_size_df
    Features,Labels=create_labels(Logs_Faults,'Events')
    # Save Features and Labels:
    save_dataset_pickle(intermediate_path,'Labels' + TagName,Labels)
    save_dataset_pickle(intermediate_path,'Features' + TagName,Features)
```
